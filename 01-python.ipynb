# Tasks

* Plot a graph of the maximum temperature at each time
* Plot a graph of the minimum temperature at each time
* Plot a graph of the mean temperature for each sensor
* Read two files and make a 2D plot of the difference at each sensor at each time

```python
import numpy as np
import matplotlib.pyplot as plt
time_max = np.max(data,axis=0)
time_min = np.min(data,axis=0)
time_mean = np.mean(data,axis=0)
plt.plot(time_max,'k',label='Maximum')
plt.plot(time_min,'r',label='Minimum')
plt.plot(time_mean,'k--',label='Mean')
plt.legend()
```

```python
sensor_max = np.max(data,axis=1)
sensor_min = np.min(data,axis=1)
sensor_mean = np.mean(data,axis=1)
plt.plot(sensor_max,'k',label='Maximum')
plt.plot(sensor_min,'r',label='Minimum')
plt.plot(sensor_mean,'k--',label='Mean')
plt.legend()
```

```python
data1=np.loadtxt('data/temperature-01.csv', delimiter=',')
data2=np.loadtxt('data/temperature-02.csv', delimiter=',')

diff = data2-data1

plt.imshow(diff)
plt.colorbar()
```
